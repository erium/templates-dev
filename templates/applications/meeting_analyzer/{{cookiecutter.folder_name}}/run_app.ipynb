# Define information stores
Define the information stores for the summaries and the transcripts.
Create them via the UI if they do not already exist.

```
summary_store_name = "MeetingSummaries"
transcript_store_name = "MeetingTranscripts"
```

# Chatbot Setup

Configure your chatbot in `chatbot.board`.
It is recommended to give it access to both information stores defined above.

```
# THIS NEEDS TO BE SETUP BY ERIUM

# start a polybot with the selected information store
import shlex
from pathlib import Path

board_path = "./chatbot.board"
card_id = "24e65a0e-6294-4f01-b6e6-246e209199b0"
logs_dir = Path("chat_logs").resolve()
chatbotport = 8502

chatbot_command = (f"polybot -b {shlex.quote(board_path)} -q {shlex.quote(card_id)} "
                   f"-c {shlex.quote(str(logs_dir))} --port {chatbotport} -s")

!register_app "{chatbot_command}" --port "{chatbotport}" --name "Meeting_Chatbot"
!sudo supervisorctl start Meeting_Chatbot__8502__root
```

# MMA Startup

```
port = 8501

mma_cmd = (f"meeting_analyzer -b 'template.board' --port {port} --cbport {chatbotport} "
           f"--summary_store {summary_store_name} --transcript_store {transcript_store_name}")

!register_app "{mma_cmd}" --port "{port}" --workdir "." --name "Meeting_Analyzer"
!sudo supervisorctl start Meeting_Analyzer__8501__root
```
